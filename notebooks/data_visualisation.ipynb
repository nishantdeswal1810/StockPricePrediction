```python
import pandas as pd

# Load the dataset
data = pd.read_csv('../data/reliance_stock_data_processed.csv', parse_dates=['Date'])

# Display the first few rows to understand the structure of the data
print(data.head())

# Check for missing values
print(data.isnull().sum())

# Get basic statistics
print(data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Get the correlation of all features with the 'Close' column
close_corr = corr_matrix['Close'].sort_values(ascending=False)
print("Correlation with Close:\n", close_corr)
```
